## Convert to PNG

```python
import os
from PIL import Image

folder_read_path = "C:/Users/richa_0/Documents/Coral Research/Images/Original_Raw_95_TIF/"
folder_write_path = "C:/Users/richa_0/Documents/Coral Research/Code Output/"

for filename in os.listdir(folder_read_path):
    if filename.endswith('.tiff') or filename.endswith('.tif'):
        tiff_path = os.path.join(folder_read_path, filename)
        png_path = os.path.join(folder_write_path, filename.split('.')[0] + '.png')
        with Image.open(tiff_path) as img:
            img.save(png_path, 'PNG')
        print(f"Converted {filename} to PNG format.")
```

## File Renamer

```python
import os
import re
import shutil

folder_read_path = "C:/Users/richa_0/Documents/Coral Research/Images/Original_Raw_All_PNG"
folder_write_path = "C:/Users/richa_0/Documents/Coral Research/Code Output"
error_write_path = "C:/Users/richa_0/Documents/Coral Research/Code Output 1"

for image_file in os.listdir(folder_read_path):
    if image_file.endswith('.png'):
        if image_file.startswith('_'):  # Ex: _20220909_16_2 
            match = re.search(r'_(\d+)_(\d+)_(\d+)', image_file)
            if match:
                date, cell_id, coral_number = match.groups()
                new_filename = f"{cell_id}_{coral_number}_{date}.png"
            else:
                new_filename = image_file
        else:  # Ex: 16_2_0_20221123
            match = re.search(r'(\d+)_(\d+)_(\d+)_(\d+)', image_file)
            if match:
                cell_id, coral_number, remove, date = match.groups()
                new_filename = f"{cell_id}_{coral_number}_{date}.png"
            else:
                new_filename = image_file
                
        image_read_path = os.path.join(folder_read_path, image_file)
        image_write_path = os.path.join(folder_write_path, new_filename)
        if not os.path.exists(image_write_path):
            shutil.copyfile(image_read_path, image_write_path)
        else:
            image_write_path = os.path.join(error_write_path, new_filename)
            print(f"Destination file already exists: {image_write_path}")
            shutil.copyfile(image_read_path, image_write_path)
```

## Missing File Finder

```python
import os

tif_folder = "C:/Users/richa_0/Documents/Coral Research/Images/Original_Raw_All_TIF/Raw_Meg_All783"
png_folder = "C:/Users/richa_0/Documents/Coral Research/Images/Original_Raw_All_PNG"

tif_files = os.listdir(tif_folder)
png_files = os.listdir(png_folder)

print("Number of TIFF files:", len(tif_files))
print("Number of TIFF files:", len(png_files))

missing_files = []

# Iterate over the range of file counts
for count in range(0, 782):
    tif_filename_withextension = tif_files[count]
    png_filename_withextension = png_files[count]
    tif_filename, tif_fileextension = os.path.splitext(tif_filename_withextension)
    png_filename, png_fileextension = os.path.splitext(png_filename_withextension)

    # Check if the file is in the tif folder but not in the png folder
    if tif_filename != png_filename:
        missing_files.append(tif_filename)

# Print the missing files
print("Missing Files:")
for missing_file in missing_files:
    print(missing_file)
```

## Metadata Reader

```python
from PIL import Image
from PIL.ExifTags import TAGS
 
# open the image
image = Image.open("C:/Users/richa_0\Documents/Coral Research/Images/Original_Traced_TIF/_20220910_25_1_measure.tif")
 
# extracting the exif metadata
exifdata = image.getexif()
 
# looping through all the tags present in exifdata
for tagid in exifdata:
     
    # getting the tag name instead of tag id
    tagname = TAGS.get(tagid, tagid)
 
    # passing the tagid to get its respective value
    value = exifdata.get(tagid)
   
    # printing the final result
    print(f"{tagname:25}: {value}")
```
